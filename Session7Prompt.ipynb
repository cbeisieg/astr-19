```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)

def my_exp(x):
    return np.exp(x)
```

```python
y = my_exp(x)

plt.plot(x, y)

plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")

plt.show()
```
